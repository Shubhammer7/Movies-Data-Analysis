```python
df_recommendation = df_new[['title', 'genres', 'keywords', 'overview']]
df_recommendation.head()
```

```python
##Data Cleaning
#Cleaning 'keywords' and 'genres'

def clean_keywords(keywords_str):
    
    try:
        keywords_list = ast.literal_eval(keywords_str)
        
        keywords = [keyword['name'] for keyword in keywords_list]
        
        return keywords
    
    except (ValueError, SystaxError):
        
        return []

def clean_genres(genre_str):
    
    try:
        genre_list = ast.literal_eval(genre_str)
        
        genres = [genre['name'] for genre in genre_list]
        
        return genres
    
    except (ValueError, SyntaxError):
        
        return []
    
#Cleaning 'Overview Column'   

df_recommendation['overview'] = df_recommendation['overview'].fillna('')

df_recommendation['overview'] = df_recommendation['overview'].str.lower()

df_recommendation['overview'] = df_recommendation['overview'].apply(
    lambda x: x.translate(str.maketrans('', '', string.punctuation))
)



df_recommendation['genres'] = df_recommendation['genres'].apply(clean_genres)
df_recommendation['keywords'] = df_recommendation['keywords'].apply(clean_keywords)
```

```python
df_recommendation['overview'][1]
```

```python
df_recommendation['genres'] = df_recommendation['genres'].astype(str)
df_recommendation['keywords'] = df_recommendation['keywords'].astype(str)
df_recommendation['overview'] = df_recommendation['overview'].astype(str)

df_recommendation['combined_features'] = df_recommendation['genres'] + ' ' + df_recommendation['keywords'] + ' ' + df_recommendation['overview']
```

```python
df_recommendation['combined_features'][1]
```

```python
def clean_combined_features(combined_str):
    return combined_str.replace("[", "").replace("]", "").replace("'", "").strip()

df_recommendation['combined_features'] = df_recommendation['combined_features'].apply(clean_combined_features)
```

```python
df_recommendation['combined_features']
```

## Count Vectorization from scratch 

```python
import re

# Define a tokenize function that works on individual strings
def tokenize(text):
    # Convert to lowercase
    text = text.lower()
    # Use regex to extract words (tokens), removing punctuation and splitting by whitespace
    tokens = re.findall(r'\b\w+\b', text)
    return tokens
```

```python
# Initialize an empty set for the vocabulary
vocabulary = set()

# Iterate through each combined feature
for combined in df_recommendation['combined_features']:
    # Tokenize the combined feature using the updated function
    tokens = tokenize(combined)
    
    # Add each token to the vocabulary set
    for token in tokens:
        if isinstance(token, str):
            vocabulary.add(token)

# Convert the set to a list
vocabulary = list(vocabulary)
```

```python
vocabulary
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df_recommendation['combined_features'])

# Step 2: Convert the TF-IDF matrix to a DataFrame for better visualization
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names())

print(tfidf_df)
```

```python
first_movie_vector = tfidf_matrix[0].toarray()
print(first_movie_vector)
```

```python
tfidf_df.to_csv('tfidf_vectors.csv', index=False)
```

```python
non_zero_counts = (tfidf_df != 0).sum()
print(non_zero_counts)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix)

# Convert it to a DataFrame for better readability
cosine_sim_df = pd.DataFrame(cosine_sim)

# Display the cosine similarity matrix
print(cosine_sim_df)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = df_recommendation[df_recommendation['title'] == title].index[0]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df_recommendation.iloc[movie_indices]

# Example usage:
recommended_movies = get_recommendations('The Shawshank Redemption')
print(recommended_movies[['title', 'genres', 'combined_features']])
```

```python
print(get_recommendations('Inception'))
print(get_recommendations('The Godfather'))
print(get_recommendations('The Dark Knight'))
```
